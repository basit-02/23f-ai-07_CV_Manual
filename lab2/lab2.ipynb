Task-1: Drawing Ellipses 
•	Load an image of your choice. 
•	Draw at least three ellipses with different parameters (varying center, axes lengths, angles, and colors). 
•	Display the image with the drawn ellipses. 

```python
import cv2
import numpy as np

# Load image or create a black canvas if not found
image = cv2.imread('sample_image.jpg')
if image is None:
    image = np.zeros((500, 500, 3), dtype="uint8")

# 1. Blue ellipse: Horizontal, 0 degree rotation
cv2.ellipse(image, center=(250, 100), axes=(100, 50), angle=0, startAngle=0, endAngle=360, color=(255, 0, 0), thickness=2)

# 2. Green ellipse: Vertical, 90 degree rotation
cv2.ellipse(image, center=(150, 300), axes=(100, 50), angle=90, startAngle=0, endAngle=360, color=(0, 255, 0), thickness=-1) # -1 fills the shape

# 3. Red ellipse: Tilted at 45 degrees, partial draw (0 to 180 degrees)
cv2.ellipse(image, center=(350, 300), axes=(80, 40), angle=45, startAngle=0, endAngle=180, color=(0, 0, 255), thickness=3)

cv2.imshow('Task 1: Ellipses', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Task 2: Converting Ellipses to Polygons 
•	Choose an ellipse from Task 1. 
•	Use ellipse2Poly to convert the ellipse to a polygon. 
•	Display the image with the original ellipse and the polygonized version. 

```python
import cv2
import numpy as np

image = cv2.imread('sample_image.jpg')
if image is None:
    image = np.zeros((500, 500, 3), dtype="uint8")

# Parameters for the ellipse
center = (250, 250)
axes = (150, 80)
angle = 30

# Original ellipse drawn in faint gray for comparison
cv2.ellipse(image, center, axes, angle, 0, 360, (100, 100, 100), 1)

# Convert ellipse to polygon (delta determines the spacing between points; larger delta = rougher polygon)
pts = cv2.ellipse2Poly(center, axes, angle, 0, 360, delta=20)

# Draw the polygonized version in bright yellow
# Reshape pts to (N, 1, 2) which is the format cv2.polylines expects
pts = pts.reshape((-1, 1, 2))
cv2.polylines(image, [pts], isClosed=True, color=(0, 255, 255), thickness=2)

cv2.imshow('Task 2: Ellipse to Polygon', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Task 3: Filling Convex Polygons 
•	Select a convex polygon (you can use the polygonized ellipse from Task 2). 
•	Fill the convex polygon with a color of your choice. o Display the image with the filled convex polygon. 

```python
import cv2
import numpy as np

image = cv2.imread('sample_image.jpg')
if image is None:
    image = np.zeros((500, 500, 3), dtype="uint8")

# Generate polygon points (same as Task 2)
pts = cv2.ellipse2Poly(center=(250, 250), axes=(150, 80), angle=30, arcStart=0, arcEnd=360, delta=15)

# Fill the convex polygon with purple (BGR: 128, 0, 128)
cv2.fillConvexPoly(image, points=pts, color=(128, 0, 128))

cv2.imshow('Task 3: Filled Convex Polygon', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Task 6: Drawing Polylines 
•	Define a set of points to create a polyline (a shape with multiple connected line segments).
•	Draw the polyline on the image. 
•	Display the image with the drawn polyline.

```python
import cv2
import numpy as np

image = cv2.imread('sample_image.jpg')
if image is None:
    image = np.zeros((500, 500, 3), dtype="uint8")

# Define an array of (x, y) points
points = np.array([[100, 100], [200, 50], [300, 150], [400, 100], [450, 300], [200, 400]], np.int32)

# Reshape the points to the required format (N, 1, 2)
points = points.reshape((-1, 1, 2))

# Draw the polyline (isClosed=True connects the last point back to the first)
cv2.polylines(image, [points], isClosed=True, color=(0, 255, 0), thickness=4)

cv2.imshow('Task 6: Polylines', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Task 7: Putting Text on Images 
•	Choose an image. 
•	Add text with different font sizes, colors, and positions. 
•	Display the image with the added text. 

```python
import cv2
import numpy as np

image = cv2.imread('sample_image.jpg')
if image is None:
    image = np.zeros((500, 500, 3), dtype="uint8")

# Text 1: Standard Simplex Font
cv2.putText(image, text='Hello OpenCV!', org=(50, 100), fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
            fontScale=1.5, color=(255, 0, 0), thickness=2)

# Text 2: Complex Font, smaller scale
cv2.putText(image, text='Computer Vision Lab 2', org=(50, 200), fontFace=cv2.FONT_HERSHEY_COMPLEX, 
            fontScale=1, color=(0, 255, 255), thickness=1)

# Text 3: Italicized Font
cv2.putText(image, text='Drawing Functions', org=(50, 300), fontFace=cv2.FONT_HERSHEY_TRIPLEX | cv2.FONT_ITALIC, 
            fontScale=1.2, color=(0, 255, 0), thickness=2)

cv2.imshow('Task 7: Text', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Task 8: Additional Challenges 
•	Implement any two additional drawing functions. 
•	Experiment with various parameters and observe the effects. 
•	Document your findings and display the image with the applied functions. 

```python
import cv2
import numpy as np

image = cv2.imread('sample_image.jpg')
if image is None:
    image = np.zeros((500, 500, 3), dtype="uint8")

"""
Challenge 1: cv2.drawMarker
Findings: Highly useful for pinpointing specific coordinates (like the center of a detected object) 
without drawing full shapes. You can change marker types (e.g., cross, star, square).
"""
cv2.drawMarker(image, position=(250, 100), color=(0, 0, 255), markerType=cv2.MARKER_CROSS, markerSize=40, thickness=3)
cv2.drawMarker(image, position=(350, 100), color=(255, 0, 0), markerType=cv2.MARKER_STAR, markerSize=40, thickness=2)

"""
Challenge 2: cv2.fillPoly
Findings: Unlike fillConvexPoly, fillPoly can handle non-convex (concave) polygons, 
such as a star shape or an hourglass shape, and can fill multiple polygons at once.
"""
# Defining an hourglass shape (non-convex)
pts = np.array([[100, 250], [400, 250], [100, 450], [400, 450]], np.int32)
pts = pts.reshape((-1, 1, 2))

# Filling the non-convex shape
cv2.fillPoly(image, [pts], color=(150, 150, 0))

cv2.imshow('Task 8: Additional Functions', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
