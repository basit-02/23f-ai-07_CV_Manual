Lab Task 1:  
Basic Image Operations 
Instructions: 
•	Read any open-source image. 
•	Resize the image to half its original dimensions. 
•	Rotate the resized image 180 degrees clockwise. 
•	Flip the rotated image vertically. 

```python
import cv2

# 1. Read an open-source image
image = cv2.imread('sample_image.jpg')

if image is None:
    print("Error: Image not found. Please check the file path.")
else:
    # Get the original dimensions
    height, width = image.shape[:2]
    
    # Calculate half dimensions
    new_width = width // 2
    new_height = height // 2
    
    # 2. Resize the image to half its original dimensions
    resized_image = cv2.resize(image, (new_width, new_height))
    
    # 3. Rotate the resized image 180 degrees clockwise
    rotated_image = cv2.rotate(resized_image, cv2.ROTATE_180)
    
    # 4. Flip the rotated image vertically (flipCode 0 means vertical flip)
    flipped_image = cv2.flip(rotated_image, 0)
    
    # Display the final image
    cv2.imshow('Task 1: Final Output', flipped_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

Lab Task 2: 
Image Filtering Techniques 

Instructions: 
•	Take the image from Exercise 2. 
•	Blur the image using a Gaussian blur with a kernel size of 5x5. 
•	Convert the blurred image to grayscale. 
•	Save the image

```python
import cv2

# 1. Take the image
image = cv2.imread('sample_image.jpg')

if image is None:
    print("Error: Image not found. Please check the file path.")
else:
    # 2. Blur the image using a Gaussian blur with a kernel size of 5x5
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
    
    # 3. Convert the blurred image to grayscale
    gray_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)
    
    # 4. Save the image to the local filesystem
    output_filename = 'task2_output_gray_blurred.jpg'
    cv2.imwrite(output_filename, gray_image)
    print(f"Success! Processed image saved as '{output_filename}'")
    
    # Create a normal, resizable window before showing the image
    cv2.namedWindow('Task 2: Blurred and Grayscale', cv2.WINDOW_NORMAL)
    
    # Display the final image
    cv2.imshow('Task 2: Blurred and Grayscale', gray_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
