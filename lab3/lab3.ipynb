Lab Task 1: Processing Images of a Video 
Instructions: 
•	Choose a video file (e.g., 'sample_video.mp4'). o Load the video and process each frame using OpenCV. 
•	Apply at least three different image processing techniques (e.g., resizing, blurring, color manipulation) to the frames. o Display the original and processed frames side by side for visual comparison. 

```python
import cv2
import numpy as np

# Load the video file
video = cv2.VideoCapture("sample_video.mp4")

if not video.isOpened():
    print("Error: Could not open video file.")
    exit()

# Set new size for resizing
new_width, new_height = 500, 480

while True:
    ret, frame = video.read()
    if not ret:
        break

    # 1. Resize frame
    resized = cv2.resize(frame, (new_width, new_height))

    # 2. Apply Gaussian Blur
    blurred = cv2.GaussianBlur(resized, (15, 15), 0)

    # 3. Convert to grayscale (and back to BGR for stacking)
    gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
    gray_bgr = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    # Combine frames side by side for visual comparison
    combined = np.hstack((resized, blurred, gray_bgr))
    cv2.imshow("Task 1: Original | Blurred | Grayscale", combined)

    # Press 'q' to quit early
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# Release everything
video.release()
cv2.destroyAllWindows()
```

Lab Task 2: Writing to Video with OpenCV
Instructions: 
•	Extend the code from Lab Task 1. 
•	After processing each frame, write the processed frame to a new video file. 
•	Set video parameters such as frame width, height, and frames per second (fps). 
•	Display a message indicating the progress of video creation. 

```python
import cv2
import numpy as np

# Load the video file
video = cv2.VideoCapture("sample_video.mp4")

if not video.isOpened():
    print("Error: Could not open video file.")
    exit()

fps = video.get(cv2.CAP_PROP_FPS)
new_width, new_height = 500, 480 

# Get total frames for our progress tracker
total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
frame_count = 0

# Create VideoWriter to save a PROCESSED video
fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
out = cv2.VideoWriter('Task2_Processed_Output.mp4', fourcc, fps, (new_width, new_height))

while True:
    ret, frame = video.read()
    if not ret:
        break
        
    frame_count += 1

    # Apply processing
    resized = cv2.resize(frame, (new_width, new_height))
    blurred = cv2.GaussianBlur(resized, (15, 15), 0)
    gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
    gray_bgr = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    # Combine frames side by side for display
    combined = np.hstack((resized, blurred, gray_bgr))
    cv2.imshow("Task 2: Processing and Saving", combined)

    # Write the PROCESSED frame (blurred) to output video
    out.write(blurred)
    
    # Print progress to the console every 30 frames
    if frame_count % 30 == 0:
        print(f"Progress: Processed {frame_count} / {total_frames} frames...")

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# Release everything
video.release()
out.release()
cv2.destroyAllWindows()

print("Task 2 Complete! Processed video saved as 'Task2_Processed_Output.mp4'")
```

Lab Task 3: Writing Text on Video
Instructions: 
•	Extend the code from Lab Task 2. 
•	Write dynamic text on each frame, indicating the frame number or any relevant information. 
•	Choose a distinctive font, size, and color for the text. 
•	Display a message indicating the completion of the video creation process. 

```python
import cv2
import numpy as np

# Load the video file
video = cv2.VideoCapture("sample_video.mp4")

if not video.isOpened():
    print("Error: Could not open video file.")
    exit()

fps = video.get(cv2.CAP_PROP_FPS)
new_width, new_height = 500, 480 

total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
frame_count = 0

# Create VideoWriter for the annotated video
fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
out = cv2.VideoWriter('Task3_Text_Output.mp4', fourcc, fps, (new_width, new_height))

while True:
    ret, frame = video.read()
    if not ret:
        break
        
    frame_count += 1

    # Apply processing
    resized = cv2.resize(frame, (new_width, new_height))
    blurred = cv2.GaussianBlur(resized, (15, 15), 0)
    gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
    gray_bgr = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)
    
    # --- TASK 3 ADDITION ---
    # Add dynamic text to the processed frame (yellow color, distinctive font)
    text = f"Frame: {frame_count} / {total_frames}"
    cv2.putText(blurred, text, (20, 40), cv2.FONT_HERSHEY_DUPLEX, 1.2, (0, 255, 255), 2)

    # Combine frames side by side for display
    combined = np.hstack((resized, blurred, gray_bgr))
    cv2.imshow("Task 3: Text Overlay Processing", combined)

    # Write the annotated frame to output video
    out.write(blurred)
    
    if frame_count % 30 == 0:
        print(f"Progress: Processed {frame_count} / {total_frames} frames...")

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# Release everything
video.release()
out.release()
cv2.destroyAllWindows()

print("Task 3 Complete! Video with text overlay successfully saved as 'Task3_Text_Output.mp4'")
```
