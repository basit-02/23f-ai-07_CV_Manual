Lab Task 1:  
Advanced Edge Detection Techniques 
Instructions: 
•	Choose an image for testing. 
•	Implement at least two additional edge detection techniques (e.g., Sobel, Scharr, Laplacian) using OpenCV. 
•	Display the resulting images after applying each edge detection technique. 
•	Provide a comparative analysis of the different edge detection methods used. 

```python
import cv2
import numpy as np

# Load image or generate a synthetic one if missing
image = cv2.imread('sample_image.jpg')
if image is None:
    print("Image not found. Generating a synthetic test image with edges...")
    image = np.zeros((400, 400, 3), dtype=np.uint8)
    cv2.rectangle(image, (100, 100), (300, 300), (255, 255, 255), -1)
    cv2.circle(image, (200, 200), 50, (128, 128, 128), -1)

# Convert to grayscale (Edge detection works best on single-channel images)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 1. Sobel Edge Detection (Combining X and Y gradients)
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.convertScaleAbs(sobel_x) + cv2.convertScaleAbs(sobel_y)

# 2. Scharr Edge Detection (More sensitive than Sobel)
scharr_x = cv2.Scharr(gray, cv2.CV_64F, 1, 0)
scharr_y = cv2.Scharr(gray, cv2.CV_64F, 0, 1)
scharr_combined = cv2.convertScaleAbs(scharr_x) + cv2.convertScaleAbs(scharr_y)

# 3. Laplacian Edge Detection (Second derivative, detects all directions at once)
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian_abs = cv2.convertScaleAbs(laplacian)

# Resize for display side-by-side
dim = (300, 300)
disp_gray = cv2.resize(gray, dim)
disp_sobel = cv2.resize(sobel_combined, dim)
disp_scharr = cv2.resize(scharr_combined, dim)
disp_laplacian = cv2.resize(laplacian_abs, dim)

# Add labels for clarity
cv2.putText(disp_gray, "Original Gray", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
cv2.putText(disp_sobel, "Sobel", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
cv2.putText(disp_scharr, "Scharr", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
cv2.putText(disp_laplacian, "Laplacian", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

# Combine images horizontally
row1 = np.hstack((disp_gray, disp_sobel))
row2 = np.hstack((disp_scharr, disp_laplacian))
final_display = np.vstack((row1, row2))

# Print Comparative Analysis
print("\n--- COMPARATIVE ANALYSIS ---")
print("1. Sobel: Good for general edge detection. Uses a 3x3 kernel but can miss finer details.")
print("2. Scharr: Similar to Sobel but uses different weights. It captures higher frequency edges and is more sensitive to subtle gradients.")
print("3. Laplacian: Uses the second derivative. It highlights regions of rapid intensity change in all directions simultaneously but is highly sensitive to noise (which is why blurring first is crucial).")
print("----------------------------\n")

cv2.imshow('Task 1: Advanced Edge Detection', final_display)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Instructions: 
•	Select an image for testing. 
•	Implement a custom blurring filter using convolution operations (you can use cv2.filter2D). 
•	Display the image after applying the custom blurring filter. o Experiment with different kernel sizes and shapes. 
•	Compare the results with the standard blurring techniques used in the previous lab activity. 

```python
import cv2
import numpy as np

# Load image or generate a synthetic one if missing
image = cv2.imread('sample_image.jpg')
if image is None:
    print("Image not found. Generating a synthetic test image...")
    image = np.zeros((400, 400, 3), dtype=np.uint8)
    # Add noise to simulate real-world conditions requiring blur
    cv2.randn(image, (128, 128, 128), (50, 50, 50))
    cv2.rectangle(image, (100, 100), (300, 300), (0, 0, 255), -1)

# 1. Create a 5x5 custom averaging kernel
# A 5x5 matrix of ones, divided by 25 to normalize the weights (so the image doesn't blow out to white)
kernel_5x5 = np.ones((5, 5), np.float32) / 25.0
custom_blur_5 = cv2.filter2D(image, -1, kernel_5x5)

# 2. Create a much stronger 15x15 custom averaging kernel
kernel_15x15 = np.ones((15, 15), np.float32) / 225.0
custom_blur_15 = cv2.filter2D(image, -1, kernel_15x15)

# 3. Standard Gaussian Blur for comparison (5x5)
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)

# Resize for clean side-by-side display
dim = (300, 300)
disp_orig = cv2.resize(image, dim)
disp_cust_5 = cv2.resize(custom_blur_5, dim)
disp_cust_15 = cv2.resize(custom_blur_15, dim)
disp_gauss = cv2.resize(gaussian_blur, dim)

# Add labels
cv2.putText(disp_orig, "Original", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
cv2.putText(disp_cust_5, "Custom 5x5 Filter", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
cv2.putText(disp_cust_15, "Custom 15x15 Filter", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
cv2.putText(disp_gauss, "Gaussian Blur 5x5", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

# Combine images into a 2x2 grid
row1 = np.hstack((disp_orig, disp_cust_5))
row2 = np.hstack((disp_cust_15, disp_gauss))
final_display = np.vstack((row1, row2))

print("\n--- KERNEL COMPARISON ---")
print("1. Custom 5x5 Filter: Applies a basic box blur. All pixels in the 5x5 grid are weighted equally.")
print("2. Custom 15x15 Filter: Uses a larger kernel. The blur is significantly stronger, wiping out smaller details.")
print("3. Gaussian Blur: Unlike the custom box filter, it uses a bell-curve weighting. Pixels closer to the center matter more, resulting in a 'smoother' and more natural-looking blur compared to the harsher filter2D box blur.")
print("-------------------------\n")

cv2.imshow('Task 2: Custom Filtering', final_display)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
